```python
from docx import Document
import os
import numpy as np
import pandas as pd
```

```python
raw_notes = Document('../data/notes.docx')
```

```python
def split_the_document(doc):
    """
    Splits the document into sections based on Heading 3 paragraphs.
    Returns a list of tuples (heading, content).
    """
    sections = []
    current_heading = None
    current_content = []

    for paragraph in doc.paragraphs:
        paragraph_text = paragraph.text.strip()

        if paragraph.style.name.startswith('Heading 3'):
            if current_heading is not None:
                sections.append((current_heading, '\n'.join(current_content).strip()))
            current_heading = paragraph_text
            current_content = []
        elif current_heading is not None and paragraph_text:
            current_content.append(paragraph_text)

    if current_heading is not None:
        sections.append((current_heading, '\n'.join(current_content).strip()))

    return sections


sections = split_the_document(raw_notes)
```

```python
print("Sections extracted from the document:")
for heading, content in sections:
    print(f"Heading: {heading}")
    print(f"Content: {content[:100]}...")  # Print first 100 characters of content
    print("-" * 40)
```

```python
import pandas as pd

# Keep the section-level table separate from later export tables.
df_sections = pd.DataFrame(sections, columns=['Heading', 'Content'])

def record_mapping_for_each_focus_group(df):
    mapping = {}
    for _, row in df.iterrows():
        heading = row['Heading']
        content_lines = [line.strip() for line in row['Content'].split('\n') if line.strip()]
        if heading not in mapping:
            mapping[heading] = []
        mapping[heading].extend(content_lines)
    return mapping
```

```python
import json

record_mapping = record_mapping_for_each_focus_group(df_sections)
print(json.dumps(record_mapping, indent=2, ensure_ascii=False))
```

```python
#get first two values from list of the keys in the record_mapping dictionary
first_two_values = {key: value[:2] for key, value in record_mapping.items()}
print(json.dumps(first_two_values, indent=2, ensure_ascii=False))
```

```python
#drop the first two values from the list of the keys in the record_mapping dictionary
records = {key: value[2:] for key, value in record_mapping.items()}
print(json.dumps(records, indent=2, ensure_ascii=False))
```

```python
flattened_data = [
    (key, value_index, value)
    for key, values in records.items()
    for value_index, value in enumerate(values)
]
df_export = pd.DataFrame(flattened_data, columns=['Focus Group', 'Value Index', 'Content'])
df_export
```

```python
# save the traceable export dataframe to a new CSV file
df_export.to_csv("record_mapping.csv", index=False)
```

```python
def find_using_keywords(keywords, required_word=None):
    if not isinstance(keywords, list):
        raise ValueError("Keywords should be provided as a list.")

    if required_word is not None and not isinstance(required_word, str):
        raise ValueError("required_word should be a string or None.")

    def matches(text):
        text = str(text).lower()
        keywords_match = any(keyword.lower() in text for keyword in keywords)
        required_match = True if required_word is None else required_word.lower() in text
        return keywords_match and required_match

    mask = df_export['Content'].apply(matches)
    filtered_df = df_export[mask]
    return filtered_df
```
